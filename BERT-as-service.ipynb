```python
questions = dataset['Question'].to_numpy().astype(str).tolist()
label = dataset['labels']
```

```python
# Pirmo 10 jautājumu garumi
for i in range(10):
    print(len(questions[i]))
```

```python
#BERT Multilingual Cased
from bert_serving.client import BertClient
import datetime
date_start = datetime.datetime.now()
print(date_start)
bc = BertClient()
encoded_questions = bc.encode(questions)
date_end = datetime.datetime.now()
print(len(encoded_questions[0]))
encoded_dataset = pd.DataFrame(encoded_questions)
encoded_dataset['labels'] = label
encoded_dataset.to_csv('BERT_multi_cased_embeddings.csv')
time_elapsed = date_end - date_start
print(date_end)
print('Time elapsed: ', time_elapsed)
```
